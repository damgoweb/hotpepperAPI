```python
import requests
import pandas as pd
```

```python
# API設定
URL = 'https://webservice.recruit.co.jp/hotpepper/gourmet/v1/'
API_KEY = 'c5c2bf7580ab2166'
query = {
    'key':API_KEY,
#'large_area':'Z012',
    'genre':'G007',
    'middle_area':['Y855','Y856','Y857'],
    'range':5,
    'format':'json',
    'count':100
}
```

```python
# APIからデータを取得
response = requests.get(URL,query)
```

```python
print(response.text)
```

```python
#レスポンスをjson変換
json_data = response.json()
json_data
```

```python
stores=""
stores= json_data['results']['shop']
```

```python
df = pd.json_normalize(stores,sep='_')
df
```

```python
df.to_csv('json_normalize3.csv',index=False,columns=["id","name","lat","lng"])
```
